```python
# Function to fetch engilsh google news from the past 30 days.
def getGoogleNews(searchterm, page):
    googlenewsobj = GoogleNews(lang='en')
    googlenewsobj = GoogleNews(period='30d')
    googlenewsobj.search(searchterm)
    newsresult = googlenewsobj.page_at(page)
    return newsresult
```

```python
# Calling getGoogleNews method to fetch the news from Google news API.
fnewsarray = getGoogleNews('Startups', 1)
snewsarray = getGoogleNews('Startups', 2)
newsarray = fnewsarray + snewsarray
print(len(newsarray), " Startups news fetched.")
```

```python
# Looping through the fetched news to visualize it clearly.
for news in newsarray:
    print("Title of news : ", news["title"])
    print("Media source : ", news["media"])
    print("Published time : ", news["date"])
    print("Published datetime : ", news["datetime"])
    print("News Description : ", news["desc"])
    print("News link : ", news["link"])
    print("News Image : ", news["img"])
    print("-----------------------------------------------------------------------------------------------------------------------------")
```

```python
# Function to return all the news titles.
def getallnewstitles(newsarray):
    newstitles = ''
    for news in newsarray:
        newstitles += news["title"]
    return newstitles

newstitles = getallnewstitles(newsarray)
newstitles
```

```python
# Importing nltk and downloading all the required NLTK resources for part-of-speech tagging,
# named entity recognition, and word data.
import nltk

nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('punkt')
nltk.download('words')
```

```python
# Importing all the required nltk modules for named entity recognition and tokenization.
from nltk.tree import Tree
from nltk import ne_chunk, pos_tag, word_tokenize
```

```python
# Function to return the named entity chunked news titles.
def getnechunkednewstitles(newstitles):
    # tokenizing, tagging and then chunking all the newstitles
    newstitletoken = word_tokenize(newstitles)
    postaggedtoken = pos_tag(newstitletoken)
    nechunked_newstitles = ne_chunk(postaggedtoken)
    return nechunked_newstitles
nechunked_newstitles = getnechunkednewstitles(newstitles)
nechunked_newstitles
```

```python
# Initializing entitycounts dictionary to store entity counts.
entitycounts = {
    "DATE": 0,
    "FACILITY": 0,
    "GPE": 0,
    "LOCATION": 0,
    "MONEY": 0,
    "ORGANIZATION": 0,
    "PERCENT": 0,
    "PERSON": 0,
    "TIME": 0
}

# Function to find and count the entity present in news titles
def findandcountenitity(nechunked_newstitles):
    # looping through the named entity chunked news titles and checking the type of entity and then counting the present of it.
    for nechunked_newstitle in nechunked_newstitles:
        if(type(nechunked_newstitle) == Tree):
            entityname = ""
            for nechunked_newstitle_leaf in nechunked_newstitle.leaves():
                entityname += nechunked_newstitle_leaf[0]
                print("| Entity Type: " + nechunked_newstitle.label(), "| Entity Name: " + entityname )
                print("|------------------------------------------------------------------------------------------|")


                if(nechunked_newstitle.label() in entitycounts):
                    entitycounts[nechunked_newstitle.label()] += 1
```

```python
# Calling function to find and count the entity present in news titles
findandcountenitity(nechunked_newstitles)
```

```python
# Printing the count of presence of each type of entities.
print("-----------Counts of entities:-----------------")
for entity in entitycounts:
    print("| " + entity + " COUNT: ", entitycounts[entity])
    print("|--------------------------|----------------------------|")
```
